## 1. My lane as an ML task (type)

**Target Variable:** `did_book_discovery_call` (Binary: 1 or 0)

## 3. Success metric

*One metric you can defend. What number means 'good'?*

**Success Metric:** Precision 
**Target Value:** 80% Precision

**Defense of Metric:**
In our B2B pipeline, a False Positive occurs when the model falsely predicts a lead will book a call, triggering personalized, resource-heavy automated workflows for an unqualified target. This costs our operations team valuable time and manual follow-up. By optimizing for 80% Precision, we ensure that the leads we target are highly qualified, minimizing wasted operational effort and protecting our domain sender reputation.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

**Why Machine Learning beats a static Python if-statement:**

1. **Messy and Unstructured Web Data:** Scraped technology indicators are highly inconsistent. A fixed rule (like checking for the exact keyword 'react') fails if a site hides its framework or uses a synonym. An ML classifier handles fuzzy text patterns seamlessly.
2. **Non-linear Signals:** No single feature determines a high-value lead. An ML model can weigh dozens of weak signals collectively (such as employee range, tech stack density, and domain type) to find hidden patterns a manual if-statement would miss.
3. **Adaptability over Time:** As business targets change and web technologies evolve, we can simply retrain our model on new observed labels rather than manually rewriting complex, brittle conditional rules.
